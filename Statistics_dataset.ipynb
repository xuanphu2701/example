```python
import pandas as pd
import numpy as np
from glob import glob
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Doc dataset va them cot domain

domains = []
file_list = glob("*.xlsx")
df = pd.DataFrame()
for i in range(len(file_list)):
  data = pd.read_excel(file_list[i])
  domain = re.findall(r'cmt_(.*).xlsx', file_list[i])
  domains.append(domain[0])
  data['domain'] = [domain[0] for i in range(len(data))]
  df = pd.concat([df, data], axis = 'index')
```
